```python
import matplotlib.pyplot as plt
import random 
```

```python

def plotHeadsProbabilityFunction(finalIteration):
  iterations = 1
  probability=[]
  flipNumbers=[]

  while iterations<finalIteration:
      head=0
      tail=0

      for i in range(iterations):
          if random.randint(0,1)==0:
              head+=1
          else:
              tail+=1
      k=head/(head+tail)
      probability.append(k)
      flipNumbers.append(iterations)
      iterations+=1

  plt.subplot(2,1,1)
  plt.hist(probability,100,label='Probality of Heads')
  plt.legend()


  plt.subplot(2,1,2)
  plt.plot(flipNumbers,probability)
  plt.xlabel('Number of Flips')
  plt.ylabel('Probability of Heads')
  plt.grid(True)

  plt.show()
```

```python
plotHeadsProbabilityFunction(2000)
```

```python
plotHeadsProbabilityFunction(200)
```
