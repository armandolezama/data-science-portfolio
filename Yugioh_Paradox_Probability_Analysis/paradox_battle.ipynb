# Paradox battle description

**[Paradox](https://yugioh.fandom.com/wiki/Paradox)** is a charater from the movie **[Yu-gi-oh! 3D: Bonds Beyond Time](https://en.wikipedia.org/wiki/Yu-Gi-Oh!_Bonds_Beyond_Time)**. He confront the **three main characters** from the **three first yugioh series** (Yugioh, Yugioh GX and Yugioh 5D) and duel links takes advantaje from this movie creating a **kind of triple duel** with certain **rulez and skills**.

**[Duel links](https://www.konami.com/yugioh/duel_links/en/)** is a game based on [speed duels](https://img.yugioh-card.com/uk/gameplay/detail.php?id=5618#:~:text=Speed%20Dueling%20is%20a%20specialised,Cards%2C%20and%20battle%20your%20opponent.), this means that there are only three slots for monster and three for spells/traps, and two extra slots for link monsters (one player can use only one slot, and can set links, fusion, sinchro or xyz monsters). In the game are playable characters, and can be used with their skills, some characters are capable to set two extra slots for pendulum cards, but this last feature is irrelevant for the purpose of this analysis.

This game use events, and fight against paradox is one of them. The event have certain [features](https://www.duellinksmeta.com/articles/events/november-2022/paradox-event), for example, some monsters can do attacks with multiplied damaged (x5), and the boss (paradox) have thousands of life points, reaching even 5 hundred thousand (it's the bigger life pints that i've seen). 

The fight against paradox includes another advantajes, for example, after complete 5 duels against the boss, you can play a [friendship duel](https://yugipedia.com/wiki/Raid_Duel_-_Bonds_Beyond_Time:_Paradox_vs._The_Three_Duelists#Friendship_Duel) which bring habilities really useful, for example, summon (without requirements) black magician, elemental hero neos, or [Stardust Dragon](https://yugioh.fandom.com/wiki/Stardust_Dragon). You can use this habilities 4 times. The last hability is about sum the attack points of two monsters to another, and then double it's attack. Using the shooting star dragon, and it's hability, you can make a lot of damage.

The last importan feature is one item used to draw two cards each turn instead one. This item, I gues is the key to get the most quantity of damage.

## Key cards

**[Shooting Star Dragon](https://www.db.yugioh-card.com/yugiohdb/card_search.action?ope=2&cid=9117)**: This card allow to make attacks equal to the tunner monsters founded when you excavate the first 5 cards on the top of the deck (this means, this card can make a maximum of 5 attacks, if you're lucky). This card is only available using [Yusei Fudo skill *Shooting Star Road*](https://yugioh.fandom.com/wiki/Yusei_Fudo_(Duel_Links)#Skills).

**[Junk Synchron](https://yugioh.fandom.com/wiki/Junk_Synchron)**: This card is a tunner monster, and using Yusei skills, is capable to summon the Shooting Star Dragon.

**[Power of the guardians](https://yugioh.fandom.com/wiki/Power_of_the_Guardians)**: this is an equip spell. Each time the equipped monster is involved in any attack, the spell places a counter in the card. The equipped monster gains 500 attack points for each spell counter.

**[Counter gate](https://www.db.yugioh-card.com/yugiohdb/card_search.action?ope=2&cid=12493)**: This trap negate one attack from opponent's monsters, and then, allows to the player to draw a card. If it's a monster, you can normal summon in the oponnent's turn.

**[Reload](https://yugipedia.com/wiki/Reload)**: This card makes the player return it's cards to the deck (except the spell card used) and then draw the same quantity of cards.

**[Magical Mallet](https://yugioh.fandom.com/wiki/Magical_Mallet)**: this card allows the player to select cards in it's hand and then return them to the deck, then draw the same quantity. 

**Tuner monsters**: This monsters, doesn't matter really what are them (except for junk synchorn, which is the key to invoke Stardust Dragon), the important thing is have them in the top of the deck to feed the attacks of the **Shooting Star Dragon**.


## Best Scenario

**Friendship duel**
In the first player turn, the best thing that can happen is: draw three Power of the guardians, and two magic cards. Then use the duel skill to summon stardust dragon, then use the skill of Yusei to summon Shooting Star Dragon (SSD). When this dragons uses it's hability, the best result is find 5 tuner monsters, make 5 attacks.

1st attack: 
SSD attack points, with equip spell, increases from 3,300 to 4,800.
Making an attack of 24,000.
Cumulative damage is: 24,000.

2nd attack: SSD attack points, from previous to current, increases from 4,800 to 6,300, 
Making an attack of 31500.
Cumulative damage is: 24,000.

```python
import pandas as pd

def calculate_damage_per_attack():

  damage_per_turn_list = []

  equip_increment = 1500

  SSD_original_attack = 3300

  total_turns = 5

  attacks_per_turn = 5
  
  for i in list(range(1, (total_turns * attacks_per_turn) + 1)):
    ssd_attack_points = SSD_original_attack + (equip_increment * i)
    damage_per_turn_list.append([ssd_attack_points, ssd_attack_points * 5])
  
  damage_per_turn = pd.DataFrame(damage_per_turn_list)

  damage_per_turn.columns = ['SSD attack points', 'Total damage this turn']
  
  return damage_per_turn

estimated_damage = calculate_damage_per_attack()

# estimated_damage['Total damage this turn'].cumsum()

estimated_damage
```
