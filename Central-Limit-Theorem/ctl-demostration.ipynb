This work have the purpose to show a little demostration of Central Limit Theorem. 

In first place, let's describe this theorem a bit.

This theorem asserts that, when independent random variables are summed up, their properly normalized sum tends toward a normal distribution, even if the original variables themselves are not normally distributed.

Example substracted from here:
https://www.geeksforgeeks.org/python-central-limit-theorem/#:~:text=The%20definition%3A,from%20which%20we%20are%20sampling.&text=are%20the%20mean%20and%20standard%20deviation%20of%20the%20population%20respectively.

```python

import numpy
import matplotlib.pyplot as plt
 
# number of samples
sample_sizes = [1, 10, 50, 1000] 
# list of sample means
means = []

low_limit = -40

high_limit = 40

samplling_iterations = 1000

seed = 5
 
# Generating 1, 10, 30, 100 random numbers from -40 to 40
# taking their mean and appending it to list means.
for j in sample_sizes:
    # Generating seed so that we can get same result
    # every time the loop is run...
    numpy.random.seed(seed)
    sampling_means = [numpy.mean(
        numpy.random.randint(
            low=low_limit, high=high_limit, size=j)) for _i in range(samplling_iterations)]
    means.append(sampling_means)
sample_size = 0
 
# plotting all the means in one figure
fig, ax = plt.subplots(2, 2, figsize =(16, 12))
for row in range(0, 2):
    for column in range(0, 2):
        # Histogram for each x stored in means
        ax[row, column].hist(means[sample_size], 10, density = True)
        ax[row, column].set_title(label = f'The sample size here is {sample_sizes[sample_size]}')
        sample_size = sample_size + 1
plt.show()

```
